```python
#extract data from website
url = 'https://newsapi.org/v2/top-headlines?'
#providing the query parameters
parameters = {
    #'q': '', # query phrase
    'pageSize': 100,  # maximum is 100
    'apiKey': secret, # your own API key
    'sortBy':'popularity',
    'country':'jp'
}
# Call API using get method
response = requests.get(url, params=parameters)
# Convert the response to JSON format and pretty print it
response_json = response.json()
```

```python
response_json
```

```python
Description = []
authors = []

for i, article in enumerate(response_json["articles"]):
    print("Extracting Data from URL - %s"%article["url"])
    content_li = article["description"]
    content = content_li
    Description.append(content)
    authors.append("authors")
```

```python
rdf = pd.DataFrame(list(zip(authors, Description)), columns=["Label", "Description"])
```

```python
import spacy

# Load the Japanese model
nlp = spacy.load("ja_core_news_sm")
```

```python
# Tokenize the text in the DataFrame
def tokenize_text(text):
    if text:
        doc = nlp(text)
        return [token.text for token in doc]
    else:
        return []

# Apply tokenization to the "text" column
rdf['tokens']= rdf['Description'].apply(tokenize_text)
```

```python
fdf = rdf.dropna()
fdf.isnull().sum()
```

```python
from deep_translator import GoogleTranslator, LingueeTranslator

def translate_japanese_to_english(text):
    translated = GoogleTranslator('ja', 'en').translate(text)
    return translated
```

```python
fdf['translated']= fdf['Description'].apply(translate_japanese_to_english)
```

```python
import nltk
from nltk.tokenize import word_tokenize

def translate_tokens(text):
    #print(text)
    list_of_tokens = word_tokenize(text)
    #print(list_of_tokens)
    word = GoogleTranslator('ja', 'en').translate_batch(list_of_tokens)
    return word
```

```python
fdf['tokens_trans'] = fdf['Description'].apply(translate_tokens)
```
